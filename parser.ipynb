```python
from bs4 import BeautifulSoup
from IPython.display import display, HTML
import pandas as pd 
from datetime import datetime, timedelta


from __future__ import print_function

import os.path
import os 
from google.auth.transport.requests import Request
from google.oauth2.credentials import Credentials
from google_auth_oauthlib.flow import InstalledAppFlow
from googleapiclient.discovery import build
from googleapiclient.errors import HttpError
from google.auth.exceptions import RefreshError
```

```python
with open("timetable.html") as fp:
    soup = BeautifulSoup(fp)
```

```python
tables = soup.find_all('table')
del tables[0]
```

```python
titles = ([x.contents[0] for x in soup.find_all('h3')])
print(titles)
titles_val = ([index for index, x in enumerate(soup.find_all('h3'))])
print(titles_val)
```

```python
tables = [pd.read_html(str(table))[0].assign(Day=titles[index]).assign(Modifier=titles_val[index]) for index, table in enumerate(tables)]
results=pd.concat(tables)
display(results)
```

```python
results.drop(['Weeks', 'Activity Code', 'Staff'], axis=1, inplace=True)
results["Type"] = [x.split(' ')[0] for x in results["Type"]]
```

```python
display(results)
```

```python
monday = soup.find_all('h1')[0].contents[0].split(' ')[-1]
#print(datetime.strptime(date, "%d/%m/%Y"))
from datetime import datetime

starts = []
ends = []

modifier = "%Y-%m-%dT%H:%M:00"

for index, row in results.iterrows():
    start_time = row['Start Time']
    #row['start'] = ('T').join(str(datetime.strptime(monday+start_time, "%d/%m/%Y%H:%M")+ timedelta(days=row['Modifier'])).split(' '))
    #row['start'] = 
    current_start = (datetime.strptime(monday+start_time, "%d/%m/%Y%H:%M")+ timedelta(days=row['Modifier']))
    starts.append(current_start.strftime(modifier))
    appointment_len=int(row['Duration'].split(' ')[0])
    ends.append((current_start+timedelta(hours=appointment_len)).strftime(modifier))
    #print(row['start'])#.strftime
    
results['Start'] = starts
results['End'] = ends
```

```python
results.drop('Modifier', axis=1, inplace=True)
display(results)
```

```python


# If modifying these scopes, delete the file token.json.
SCOPES = ['https://www.googleapis.com/auth/calendar']


def main():
    """Shows basic usage of the Google Calendar API.
    Prints the start and name of the next 10 events on the user's calendar.
    """
    creds = None
    # The file token.json stores the user's access and refresh tokens, and is
    # created automatically when the authorization flow completes for the first
    # time.
    if os.path.exists('token.json'):
        creds = Credentials.from_authorized_user_file('token.json', SCOPES)
    # If there are no (valid) credentials available, let the user log in.
    if not creds or not creds.valid:
        if creds and creds.expired and creds.refresh_token:
            creds.refresh(Request())
        else:
            flow = InstalledAppFlow.from_client_secrets_file(
                'credentials.json', SCOPES)
            creds = flow.run_local_server(port=0)
        # Save the credentials for the next run
        with open('token.json', 'w') as token:
            token.write(creds.to_json())

    try:
        service = build('calendar', 'v3', credentials=creds)
        
        for index, row in results.iterrows():
        
            event = {
              'summary': f'{row["Type"]} - {row["Activity Name"]}',
              'location': row['Room'],
              'description': f'{row["Duration"]} {row["Activity Name"]} ({row["Module"]}) {row["Type"]} at {row["Room"]}',
              'start': {
                'dateTime': row['Start'],#-07:00',
                'timeZone': 'Europe/London',
              },
              'end': {
                'dateTime': row['End'],#-07:00',
                'timeZone': 'Europe/London',
              },
              'reminders': {
                'useDefault': False,
                'overrides': [
                  {'method': 'popup', 'minutes': 45},
                  {'method': 'popup', 'minutes': 5},
                ],
              },
            }

            event = service.events().insert(calendarId='primary', body=event).execute()
            print('Event created: %s' % (event.get('htmlLink')))
        
    except HttpError as error:
        print('An error occurred: %s' % error)

try:
    main()
except RefreshError:
    os.remove("token.json")
    main()
```
